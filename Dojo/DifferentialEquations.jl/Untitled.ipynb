```julia
function lorentz(t,u,du)
    du[1] = 10.0(u[2] - u[1])
    du[2] = u[1]*(28.0 - u[3]) - u[2]
    du[3] = u[1]*u[2] - (8/3)*u[3]
end
```

```julia
u0 = [1.0; 0.0; 0.0]
tspan = (0.0,1.0)
prob = ODEProblem(lorentz,u0,tspan)
sol = solve(prob)
```

```julia
plot(sol)
```
